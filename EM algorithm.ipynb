```python
import numpy as np
import numpy.random as rnd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
z_list = [0.2, 0.5, 0.3]

rnd.seed(18)

n_sample = 500

mu_list = stats.uniform(loc=-3, scale=6).rvs(size=(3, 2))
cov_list = stats.invwishart(df=3, scale=np.eye(2)).rvs(3)

data = []
for z, mu, cov in zip(z_list, mu_list, cov_list):
    y = stats.multivariate_normal(mean=mu, cov=cov).rvs(int(n_sample * z))
    data.append(y)
```

```python
for i, y in enumerate(data):
    plt.scatter(y[:,0], y[:,1], s=20, label=f'cluster {i}')
plt.legend()
```

```python
D = np.vstack(data)
plt.scatter(D[:,0], D[:,1], s=20)
```
